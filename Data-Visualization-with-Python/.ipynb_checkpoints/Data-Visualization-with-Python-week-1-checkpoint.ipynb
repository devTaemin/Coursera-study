### 1-2. Introduction to Matplotlib

-----------------------------------------------------------------------------

- History <br>
     __Jonh Hunter__ (1958 - 2012), EEG Visualiztion tool <br>
     Analogous to Matlab Scripting Interface.<br> 
    

- Architecture <br>
    > __Backend Layer__ (FigureCanvas, Renderer, Event) <br>
    > __Artist Layer__ (Artist) <br>
    > __Scripting Layer__ (pyplot) <br>
 

+ Backend layer 
  
   > * FigureCanvas: __matplotlib.backend_bases.FigureCanvas__<br>
     => Encompases the area onto which the figure is drawn.<br>
   > * Renderer: __matplotlib.backend_bases.Renderer__<br>
     => Knows how to draw on the FigureCanvas.<br>
   > * Event: __matplotlib.backend_bases.Event__<br>
     => Handles user inputs such as keyboard states and mouse clicks.

+ Artist layer

   > - Comprised of one main object : __Artist__ <br>
   => Knows how to use the Renderer to draw on Canvas.
   > - __Title, lines, tick labels, images__, all correspond to individual artist instaces.
   > - Two types of Artist objects <br>
       1) <b>Primitive</b>: Line2D, Rectangle, Circle, and Text <br>
       2) <b>Composite</b>: Axis, Tick, and Figure <br>
       => Each composite artist may contain other composite artists as well as primitive artists. <br>

##### Example of putting the artist layer to use for generating histogram.

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas # import figureCanvas
from matplotlib.figure import Figure # import Figure artist
fig = Figure()
canvas = FigureCanvas(fig)

# create 10000 random numbers using numpy
import numpy as np
x = np.random.randn(10000)

ax = fig.add_subplot(111) # create an axes artist

ax.hist(x, 100) # generate a histogram of the 10000 numbers

# add a title to the figure and save it
ax.set_title('Normal distribution with $\mu=0, \sigma=1$')
fig.savefig('matplotlib_histogram_artist-layer.png')
```

+ Scripting Layer

   > Comprised mainly of __pyplot__ <br>
   => A scripting interface that is __lighter than the artist layer__ <br>

##### Example of putting the scripitng layer to use for generating histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
plt.hist(x, 100)
plt.title(r'Normal distribution with $\mu=0, \sigma=1$')
plt.savefig('matplotlib_histogram_pyplot.png')
plt.show()
```
